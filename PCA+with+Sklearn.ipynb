# PCA using Sklearn

```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Download the MNIST dataset from 
#https://www.kaggle.com/c/digit-recognizer/data
d0 = pd.read_csv('train.csv')
```

```python
#save label is l
l = d0['label']
#drop the label columns
d = d0.drop('label', axis = 1)
```

```python
#Pick first 15k datapoints
labels = l.head(15000)
data = d.head(15000)
```

```python
#Standardizing the data
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
sample_data = standardized_data
```

```python
#Intializing PCA
from sklearn import decomposition 
pca = decomposition.PCA()
```

```python
#new dimension = 2, transforming the data from 784 dimesnion to 2 dimension
pca.n_components = 2 
#fitting the data on PCA
pca_data = pca.fit_transform(sample_data)

print('shape of PCA reduced shape =', pca_data.shape)
```

```python
labels.shape
```

```python
import seaborn as sns
#stacking the label column with the data columnwise
pca_data = np.vstack((pca_data.T, labels)).T

pca_data.shape
#forming a datframe 
pca_df = pd.DataFrame(data = pca_data , columns =('1st Principal', '2nd Principal', 'labels'))
sns.FacetGrid(pca_df , hue = 'labels', size = 6).map(plt.scatter, '1st Principal', '2nd Principal').add_legend()
plt.show()
```

The data has been reduced from 784 dimension to 2 dimension.
